# Educational Counseling Bias Analysis
## Visualizing Demographic Biases Across Prompt Complexity Levels

This notebook analyzes bias in AI model responses to educational counseling prompts across:
- **Question categories**: Opportunity, Potential, Discipline
- **Demographics**: Race/gender, socioeconomic status, background
- **Prompt levels**: From minimal information to explicit performance metrics

```python
import pandas as pd
import numpy as np
from scipy import stats
import json
import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.lines import Line2D
from math import pi

%matplotlib inline
plt.rcParams['figure.dpi'] = 120
plt.rcParams['font.family'] = 'sans-serif'
```

```python
# Load data - adjust paths as needed
df = pd.read_csv('../results_raw.csv')

with open('../prompts/demographics.json', 'r') as f:
    demographics = json.load(f)

print(f"Total responses: {len(df):,}")
print(f"Models: {df['model'].unique().tolist()}")
print(f"Categories: {df['category'].unique().tolist()}")
print(f"Levels: {df['level'].nunique()}")
```

```python
# Configuration
level_order = [
    'minimal', 'context_only', 'vague_positive', 'vague_negative',
    'neutral_metrics', 'positive_metrics', 'negative_metrics',
    'contradict_pos_metrics', 'contradict_neg_metrics'
]

level_labels = {
    'minimal': 'Minimal', 'context_only': 'Context', 'vague_positive': 'Vague+',
    'vague_negative': 'Vague-', 'neutral_metrics': 'Neutral', 'positive_metrics': 'Positive',
    'negative_metrics': 'Negative', 'contradict_pos_metrics': 'Contra+', 'contradict_neg_metrics': 'Contra-'
}

demo_order = [
    'aa_male', 'aa_female', 'hispanic_male', 'hispanic_female',
    'asian_male', 'asian_female', 'white_male', 'white_female',
    'low_income', 'working_class', 'middle_class', 'affluent',
    'first_gen', 'immigrant'
]

demo_labels = {
    'aa_male': 'AA Male', 'aa_female': 'AA Female',
    'hispanic_male': 'Hisp Male', 'hispanic_female': 'Hisp Female',
    'asian_male': 'Asian Male', 'asian_female': 'Asian Female',
    'white_male': 'White Male', 'white_female': 'White Female',
    'low_income': 'Low Income', 'working_class': 'Working Class',
    'middle_class': 'Middle Class', 'affluent': 'Affluent',
    'first_gen': 'First Gen', 'immigrant': 'Immigrant'
}

demo_groups = {
    'aa_male': 'African American', 'aa_female': 'African American',
    'hispanic_male': 'Hispanic', 'hispanic_female': 'Hispanic',
    'asian_male': 'Asian', 'asian_female': 'Asian',
    'white_male': 'White', 'white_female': 'White',
    'low_income': 'Socioeconomic', 'working_class': 'Socioeconomic',
    'middle_class': 'Socioeconomic', 'affluent': 'Socioeconomic',
    'first_gen': 'Background', 'immigrant': 'Background'
}

group_colors = {
    'African American': '#E64B35', 'Hispanic': '#4DBBD5', 'Asian': '#00A087',
    'White': '#3C5488', 'Socioeconomic': '#F39B7F', 'Background': '#8491B4'
}

categories = ['opportunity', 'potential', 'discipline']
```

```python
# Compute results
def compute_results():
    results = []
    for category in categories:
        cat_df = df[df['category'] == category]
        for level in level_order:
            level_df = cat_df[cat_df['level'] == level]
            control = level_df[level_df['demographic_id'] == 'control']['score']
            if len(control) == 0:
                continue
            control_mean, control_std, control_n = control.mean(), control.std(), len(control)
            
            for demo_id in demo_order:
                demo = level_df[level_df['demographic_id'] == demo_id]['score']
                if len(demo) == 0:
                    continue
                demo_mean, demo_std = demo.mean(), demo.std()
                delta = demo_mean - control_mean
                se = np.sqrt(demo_std**2/len(demo) + control_std**2/control_n)
                _, p = stats.ttest_ind(demo, control)
                
                results.append({
                    'category': category, 'level': level, 'level_idx': level_order.index(level),
                    'demographic_id': demo_id, 'demographic_label': demo_labels[demo_id],
                    'demographic_group': demo_groups[demo_id],
                    'mean': demo_mean, 'delta': delta, 'se': se,
                    'ci_lower': delta - 1.96*se, 'ci_upper': delta + 1.96*se,
                    'p_value': p, 'significant': p < 0.05
                })
    return pd.DataFrame(results)

results_df = compute_results()
print(f"Computed {len(results_df)} demographic × level × category combinations")
```

---
## 2. Publication-Quality Heatmap
Shows all combinations with significance markers (•)

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 10))

colors_neg = plt.cm.Blues_r(np.linspace(0.2, 0.8, 128))
colors_pos = plt.cm.Reds(np.linspace(0.2, 0.8, 128))
cmap = LinearSegmentedColormap.from_list('diverging', np.vstack([colors_neg, colors_pos]))

for idx, category in enumerate(categories):
    ax = axes[idx]
    cat_data = results_df[results_df['category'] == category]
    
    pivot_delta = cat_data.pivot_table(index='demographic_id', columns='level', values='delta')
    pivot_sig = cat_data.pivot_table(index='demographic_id', columns='level', values='significant')
    pivot_delta = pivot_delta.reindex(index=demo_order, columns=level_order)
    pivot_sig = pivot_sig.reindex(index=demo_order, columns=level_order)
    
    im = ax.imshow(pivot_delta.values, cmap=cmap, aspect='auto', vmin=-0.5, vmax=0.5)
    
    for i in range(len(demo_order)):
        for j in range(len(level_order)):
            val = pivot_delta.iloc[i, j]
            sig = pivot_sig.iloc[i, j]
            if pd.notna(val):
                color = 'white' if abs(val) > 0.25 else 'black'
                marker = '•' if sig else ''
                ax.text(j, i, f'{val:+.2f}{marker}', ha='center', va='center', 
                       color=color, fontsize=7, fontweight='bold' if sig else 'normal')
    
    for boundary in [1.5, 3.5, 5.5, 7.5, 11.5]:
        ax.axhline(y=boundary, color='white', linewidth=2)
    
    ax.set_xticks(range(len(level_order)))
    ax.set_xticklabels([level_labels[l] for l in level_order], rotation=45, ha='right', fontsize=9)
    ax.set_yticks(range(len(demo_order)))
    ax.set_yticklabels([demo_labels[d] for d in demo_order], fontsize=9)
    ax.set_title(f'{category.upper()}', fontsize=14, fontweight='bold', pad=10)

cbar = fig.colorbar(im, ax=axes, orientation='vertical', fraction=0.02, pad=0.02)
cbar.set_label('Δ from Control', fontsize=11)
fig.suptitle('Bias by Demographic and Prompt Level (• = p < 0.05)', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

---
## 3. Trajectory Lines
Shows how bias changes across prompt complexity levels

```python
fig, axes = plt.subplots(3, 2, figsize=(16, 14))

race_demos = ['aa_male', 'aa_female', 'hispanic_male', 'hispanic_female', 
              'asian_male', 'asian_female', 'white_male', 'white_female']
ses_demos = ['low_income', 'working_class', 'middle_class', 'affluent', 'first_gen', 'immigrant']
ses_colors = {'low_income': '#d62728', 'working_class': '#ff7f0e', 
              'middle_class': '#2ca02c', 'affluent': '#1f77b4',
              'first_gen': '#9467bd', 'immigrant': '#8c564b'}

for row_idx, category in enumerate(categories):
    cat_data = results_df[results_df['category'] == category]
    
    # Race/Gender
    ax_race = axes[row_idx, 0]
    for demo_id in race_demos:
        demo_data = cat_data[cat_data['demographic_id'] == demo_id].sort_values('level_idx')
        color = group_colors[demo_groups[demo_id]]
        linestyle = '-' if 'male' in demo_id else '--'
        ax_race.plot(demo_data['level_idx'], demo_data['delta'], 
                    color=color, linestyle=linestyle, linewidth=2, 
                    marker='o', markersize=5, label=demo_labels[demo_id], alpha=0.8)
    
    ax_race.axhline(y=0, color='black', linestyle='-', linewidth=1)
    ax_race.axhline(y=0.2, color='gray', linestyle=':', linewidth=0.5, alpha=0.5)
    ax_race.axhline(y=-0.2, color='gray', linestyle=':', linewidth=0.5, alpha=0.5)
    ax_race.set_xticks(range(len(level_order)))
    ax_race.set_xticklabels([level_labels[l] for l in level_order], rotation=45, ha='right', fontsize=8)
    ax_race.set_ylabel('Δ from Control', fontsize=10)
    ax_race.set_title(f'{category.upper()} - Race/Gender', fontsize=12, fontweight='bold')
    ax_race.set_ylim(-0.5, 0.7)
    ax_race.grid(alpha=0.3)
    if row_idx == 0:
        ax_race.legend(loc='upper right', fontsize=7, ncol=2)
    
    # SES
    ax_ses = axes[row_idx, 1]
    for demo_id in ses_demos:
        demo_data = cat_data[cat_data['demographic_id'] == demo_id].sort_values('level_idx')
        ax_ses.plot(demo_data['level_idx'], demo_data['delta'], 
                   color=ses_colors[demo_id], linewidth=2,
                   marker='s', markersize=5, label=demo_labels[demo_id], alpha=0.8)
    
    ax_ses.axhline(y=0, color='black', linestyle='-', linewidth=1)
    ax_ses.axhline(y=0.2, color='gray', linestyle=':', linewidth=0.5, alpha=0.5)
    ax_ses.axhline(y=-0.2, color='gray', linestyle=':', linewidth=0.5, alpha=0.5)
    ax_ses.set_xticks(range(len(level_order)))
    ax_ses.set_xticklabels([level_labels[l] for l in level_order], rotation=45, ha='right', fontsize=8)
    ax_ses.set_ylabel('Δ from Control', fontsize=10)
    ax_ses.set_title(f'{category.upper()} - Socioeconomic/Background', fontsize=12, fontweight='bold')
    ax_ses.set_ylim(-1.0, 1.2)
    ax_ses.grid(alpha=0.3)
    if row_idx == 0:
        ax_ses.legend(loc='upper right', fontsize=8)

fig.suptitle('Bias Trajectory Across Prompt Complexity Levels', fontsize=16, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

---
## 4. Forest Plot with Confidence Intervals
Shows average bias with uncertainty

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 10))

for idx, category in enumerate(categories):
    ax = axes[idx]
    cat_data = results_df[results_df['category'] == category]
    
    demo_avg = cat_data.groupby(['demographic_id', 'demographic_group']).agg({
        'delta': 'mean', 'ci_lower': 'mean', 'ci_upper': 'mean', 'significant': 'sum'
    }).reset_index()
    demo_avg = demo_avg.set_index('demographic_id').reindex(demo_order).reset_index()
    
    for i, row in demo_avg.iterrows():
        color = group_colors[row['demographic_group']]
        ax.plot([row['ci_lower'], row['ci_upper']], [i, i], color=color, linewidth=2, alpha=0.6)
        marker = 'o' if row['significant'] >= 3 else 's'
        ax.scatter(row['delta'], i, color=color, s=80, marker=marker, zorder=5, edgecolor='black', linewidth=0.5)
    
    ax.axvline(x=0, color='black', linestyle='-', linewidth=1, alpha=0.5)
    ax.axvline(x=0.1, color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
    ax.axvline(x=-0.1, color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
    ax.set_yticks(range(len(demo_order)))
    ax.set_yticklabels([demo_labels[d] for d in demo_order], fontsize=10)
    ax.set_xlabel('Average Δ from Control', fontsize=11)
    ax.set_title(f'{category.upper()}', fontsize=14, fontweight='bold')
    ax.set_xlim(-0.6, 0.6)
    ax.invert_yaxis()
    ax.grid(axis='x', alpha=0.3)

legend_elements = [mpatches.Patch(facecolor=c, label=g) for g, c in group_colors.items()]
fig.legend(handles=legend_elements, loc='lower center', ncol=6, fontsize=10, bbox_to_anchor=(0.5, -0.02))
fig.suptitle('Average Bias with 95% CI (○ = 3+ significant levels, □ = fewer)', fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

---
## 5. Slope Chart: Minimal → Explicit Metrics
Shows how providing explicit performance data changes bias

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 10))
compare_levels = ['minimal', 'positive_metrics']

for idx, category in enumerate(categories):
    ax = axes[idx]
    cat_data = results_df[results_df['category'] == category]
    
    for demo_id in demo_order:
        demo_data = cat_data[cat_data['demographic_id'] == demo_id]
        vals = [demo_data[demo_data['level'] == l]['delta'].values[0] 
                if len(demo_data[demo_data['level'] == l]) > 0 else np.nan 
                for l in compare_levels]
        
        if not any(pd.isna(vals)):
            color = group_colors[demo_groups[demo_id]]
            ax.plot([0, 1], vals, color=color, linewidth=2, alpha=0.7)
            ax.scatter([0, 1], vals, color=color, s=60, zorder=5, edgecolor='black', linewidth=0.5)
            ax.text(1.05, vals[1], demo_labels[demo_id], va='center', fontsize=8, color=color)
    
    ax.set_xlim(-0.2, 1.5)
    ax.set_xticks([0, 1])
    ax.set_xticklabels(['Minimal\n(No metrics)', 'Positive Metrics\n(Explicit data)'], fontsize=11)
    ax.set_ylabel('Δ from Control', fontsize=11)
    ax.set_title(f'{category.upper()}', fontsize=14, fontweight='bold')
    ax.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
    ax.grid(axis='y', alpha=0.3)

fig.suptitle('How Explicit Performance Metrics Change Bias\n(Minimal Info → Positive Metrics)', 
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

## 6. Dot Matrix
Size = magnitude, Color = direction, Border = significant

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 10))

for idx, category in enumerate(categories):
    ax = axes[idx]
    cat_data = results_df[results_df['category'] == category]
    
    pivot = cat_data.pivot_table(index='demographic_id', columns='level', values='delta')
    pivot_sig = cat_data.pivot_table(index='demographic_id', columns='level', values='significant')
    pivot = pivot.reindex(index=demo_order, columns=level_order)
    pivot_sig = pivot_sig.reindex(index=demo_order, columns=level_order)
    
    for i, demo_id in enumerate(demo_order):
        for j, level in enumerate(level_order):
            val = pivot.loc[demo_id, level] if demo_id in pivot.index and level in pivot.columns else np.nan
            sig = pivot_sig.loc[demo_id, level] if demo_id in pivot_sig.index and level in pivot_sig.columns else False
            
            if pd.notna(val):
                size = abs(val) * 400 + 20
                color = '#d62728' if val < 0 else '#2ca02c'
                alpha = 0.9 if sig else 0.4
                edgecolor = 'black' if sig else 'none'
                ax.scatter(j, i, s=size, c=color, alpha=alpha, edgecolors=edgecolor, linewidths=1.5)
    
    ax.set_xticks(range(len(level_order)))
    ax.set_xticklabels([level_labels[l] for l in level_order], rotation=45, ha='right', fontsize=9)
    ax.set_yticks(range(len(demo_order)))
    ax.set_yticklabels([demo_labels[d] for d in demo_order], fontsize=9)
    ax.set_title(f'{category.upper()}', fontsize=14, fontweight='bold')
    ax.set_xlim(-0.5, len(level_order) - 0.5)
    ax.set_ylim(-0.5, len(demo_order) - 0.5)
    ax.invert_yaxis()
    ax.grid(alpha=0.2)

legend_elements = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#2ca02c', markersize=10, label='Positive Δ'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#d62728', markersize=10, label='Negative Δ'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='gray', markersize=6, label='Small effect'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='gray', markersize=12, label='Large effect'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='gray', markeredgecolor='black', 
           markeredgewidth=2, markersize=10, label='Significant (p<.05)'),
]
fig.legend(handles=legend_elements, loc='lower center', ncol=5, fontsize=10, bbox_to_anchor=(0.5, -0.02))
fig.suptitle('Bias Magnitude and Direction (Size = |Δ|, Color = Direction, Border = Significant)', 
             fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

## 7. Small Multiples
One panel per demographic showing all three categories

```python
fig, axes = plt.subplots(4, 4, figsize=(18, 16))
axes = axes.flatten()
cat_colors = {'opportunity': '#1f77b4', 'potential': '#2ca02c', 'discipline': '#d62728'}

for i, demo_id in enumerate(demo_order):
    ax = axes[i]
    
    for category in categories:
        cat_data = results_df[(results_df['category'] == category) & (results_df['demographic_id'] == demo_id)]
        cat_data = cat_data.sort_values('level_idx')
        
        ax.plot(cat_data['level_idx'], cat_data['delta'], 
               color=cat_colors[category], linewidth=2, marker='o', markersize=4, label=category.title())
        
        sig_data = cat_data[cat_data['significant']]
        ax.scatter(sig_data['level_idx'], sig_data['delta'], 
                  color=cat_colors[category], s=60, zorder=5, edgecolor='black', linewidth=1)
    
    ax.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
    ax.set_xlim(-0.5, 8.5)
    ax.set_ylim(-1.0, 1.2)
    ax.set_title(demo_labels[demo_id], fontsize=11, fontweight='bold')
    ax.set_xticks(range(len(level_order)))
    ax.set_xticklabels([])
    ax.grid(alpha=0.2)
    
    if i >= 12:
        ax.set_xticklabels([level_labels[l][:3] for l in level_order], rotation=45, ha='right', fontsize=7)
    if i % 4 == 0:
        ax.set_ylabel('Δ', fontsize=10)

for ax in axes[14:]:
    ax.axis('off')

legend_elements = [
    Line2D([0], [0], color='#1f77b4', linewidth=2, marker='o', label='Opportunity'),
    Line2D([0], [0], color='#2ca02c', linewidth=2, marker='o', label='Potential'),
    Line2D([0], [0], color='#d62728', linewidth=2, marker='o', label='Discipline'),
    Line2D([0], [0], marker='o', color='gray', markeredgecolor='black', markeredgewidth=2, 
           markersize=10, linestyle='none', label='Significant')
]
axes[14].legend(handles=legend_elements, loc='center', fontsize=12)
axes[14].set_title('Legend', fontsize=12, fontweight='bold')

fig.suptitle('Bias Trajectories by Demographic (Each Panel = One Demographic Group)', 
             fontsize=16, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

## 8. Diverging Bar Chart
Simple summary of average bias by demographic

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 8))

for idx, category in enumerate(categories):
    ax = axes[idx]
    cat_data = results_df[results_df['category'] == category]
    
    demo_avg = cat_data.groupby('demographic_id')['delta'].mean().reindex(demo_order)
    colors = [group_colors[demo_groups[d]] for d in demo_order]
    
    bars = ax.barh(range(len(demo_order)), demo_avg.values, color=colors, edgecolor='black', linewidth=0.5, height=0.7)
    
    for i, (bar, val) in enumerate(zip(bars, demo_avg.values)):
        x_pos = val + 0.02 if val >= 0 else val - 0.02
        ha = 'left' if val >= 0 else 'right'
        ax.text(x_pos, i, f'{val:+.2f}', va='center', ha=ha, fontsize=9, fontweight='bold')
    
    ax.axvline(x=0, color='black', linewidth=1)
    ax.axvline(x=0.2, color='green', linestyle='--', linewidth=0.5, alpha=0.5)
    ax.axvline(x=-0.2, color='red', linestyle='--', linewidth=0.5, alpha=0.5)
    ax.set_yticks(range(len(demo_order)))
    ax.set_yticklabels([demo_labels[d] for d in demo_order], fontsize=10)
    ax.set_xlabel('Average Δ from Control', fontsize=11)
    ax.set_title(f'{category.upper()}', fontsize=14, fontweight='bold')
    ax.set_xlim(-0.5, 0.5)
    ax.grid(axis='x', alpha=0.3)
    ax.invert_yaxis()

legend_elements = [mpatches.Patch(facecolor=c, label=g, edgecolor='black') for g, c in group_colors.items()]
fig.legend(handles=legend_elements, loc='lower center', ncol=6, fontsize=10, bbox_to_anchor=(0.5, -0.05))
fig.suptitle('Average Bias by Demographic (Across All Prompt Levels)', fontsize=16, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

## 9. Radar Chart
Bias profiles by demographic group

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 12), subplot_kw=dict(polar=True))
axes = axes.flatten()

grouped_demos = {
    'African American': ['aa_male', 'aa_female'],
    'Hispanic': ['hispanic_male', 'hispanic_female'],
    'Asian': ['asian_male', 'asian_female'],
    'White': ['white_male', 'white_female'],
    'Lower SES': ['low_income', 'working_class'],
    'Higher SES': ['middle_class', 'affluent']
}

radar_levels = ['minimal', 'context_only', 'vague_negative', 'neutral_metrics', 'positive_metrics', 'negative_metrics']
num_vars = len(radar_levels)
angles = [n / float(num_vars) * 2 * pi for n in range(num_vars)]
angles += angles[:1]

cat_colors_radar = {'opportunity': '#1f77b4', 'potential': '#2ca02c', 'discipline': '#d62728'}

for ax_idx, (group_name, demo_ids) in enumerate(grouped_demos.items()):
    ax = axes[ax_idx]
    
    for category in categories:
        cat_data = results_df[results_df['category'] == category]
        group_data = cat_data[cat_data['demographic_id'].isin(demo_ids)]
        level_avg = group_data.groupby('level')['delta'].mean()
        
        values = [level_avg.get(l, 0) for l in radar_levels]
        values += values[:1]
        
        ax.plot(angles, values, color=cat_colors_radar[category], linewidth=2, label=category.title())
        ax.fill(angles, values, color=cat_colors_radar[category], alpha=0.1)
    
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels([level_labels[l] for l in radar_levels], fontsize=8)
    ax.set_title(group_name, fontsize=12, fontweight='bold', pad=20)
    ax.set_ylim(-0.6, 0.6)
    
    if ax_idx == 0:
        ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0), fontsize=9)

fig.suptitle('Bias Profile by Demographic Group Across Prompt Levels', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

## Summary Tables

```python
# Summary table: Average bias by category and demographic group
summary = results_df.groupby(['category', 'demographic_group']).agg({
    'delta': ['mean', 'std'],
    'significant': ['sum', 'mean']
}).round(3)
summary.columns = ['Mean Δ', 'Std Δ', 'N Significant', '% Significant']
summary['% Significant'] = (summary['% Significant'] * 100).round(1)
print("Average Bias by Category and Demographic Group:")
display(summary)
```

```python
# Summary by prompt level
level_summary = results_df.groupby('level').agg({
    'delta': ['mean', lambda x: x.abs().mean()],
    'significant': ['sum', 'mean']
}).round(3)
level_summary.columns = ['Mean Δ', 'Mean |Δ|', 'N Significant', '% Significant']
level_summary['% Significant'] = (level_summary['% Significant'] * 100).round(1)
level_summary = level_summary.reindex(level_order)
print("Bias by Prompt Complexity Level:")
display(level_summary)
```
