```python
#!/usr/bin/env python3
"""
Enhanced Bias Analysis - Detailed Prompt Level Tables by Question Category
Creates tables similar to the LaTeX format showing mean, delta, and significance
for each demographic × prompt level combination, separated by question category.
"""

import pandas as pd
import numpy as np
from scipy import stats
import json
import warnings
warnings.filterwarnings('ignore')

# === LOAD DATA ===
print("Loading data...")
df = pd.read_csv('../results_raw.csv')

with open('../prompts/questions.json', 'r') as f:
    questions = json.load(f)

with open('../prompts/demographics.json', 'r') as f:
    demographics = json.load(f)

print(f"Total responses: {len(df):,}")

# === SETUP ===
# Define level order (matching the LaTeX table)
level_order = [
    'minimal', 'context_only', 'vague_positive', 'vague_negative',
    'neutral_metrics', 'positive_metrics', 'negative_metrics',
    'contradict_pos_metrics', 'contradict_neg_metrics'
]

level_labels = {
    'minimal': 'Minimal',
    'context_only': 'Context Only',
    'vague_positive': 'Vague Pos',
    'vague_negative': 'Vague Neg',
    'neutral_metrics': 'Neutral Met',
    'positive_metrics': 'Positive Met',
    'negative_metrics': 'Negative Met',
    'contradict_pos_metrics': 'Contradict+',
    'contradict_neg_metrics': 'Contradict-'
}

# Demographic order (matching the LaTeX table structure)
demo_order = [
    'aa_male', 'aa_female',
    'hispanic_male', 'hispanic_female',
    'asian_male', 'asian_female',
    'white_male', 'white_female',
    'low_income', 'working_class', 'middle_class', 'affluent',
    'first_gen', 'immigrant'
]

demo_labels = {
    'aa_male': 'African American Male',
    'aa_female': 'African American Female',
    'hispanic_male': 'Hispanic Male',
    'hispanic_female': 'Hispanic Female',
    'asian_male': 'Asian Male',
    'asian_female': 'Asian Female',
    'white_male': 'White Male',
    'white_female': 'White Female',
    'low_income': 'Low Income',
    'working_class': 'Working Class',
    'middle_class': 'Middle Class',
    'affluent': 'Affluent',
    'first_gen': 'First Generation',
    'immigrant': 'Immigrant'
}

demo_groups = {
    'aa_male': 'Race/Gender', 'aa_female': 'Race/Gender',
    'hispanic_male': 'Race/Gender', 'hispanic_female': 'Race/Gender',
    'asian_male': 'Race/Gender', 'asian_female': 'Race/Gender',
    'white_male': 'Race/Gender', 'white_female': 'Race/Gender',
    'low_income': 'Socioeconomic', 'working_class': 'Socioeconomic',
    'middle_class': 'Socioeconomic', 'affluent': 'Socioeconomic',
    'first_gen': 'Background', 'immigrant': 'Background'
}

categories = ['opportunity', 'potential', 'discipline']

def get_significance(p_value):
    """Return significance stars"""
    if p_value < 0.001:
        return '***'
    elif p_value < 0.01:
        return '**'
    elif p_value < 0.05:
        return '*'
    else:
        return ''

def format_delta(delta):
    """Format delta with sign"""
    if delta >= 0:
        return f"+{delta:.2f}"
    else:
        return f"{delta:.2f}"

# === COMPUTE DETAILED TABLES ===
print("\nComputing detailed tables by category and prompt level...")

all_results = []

for category in categories:
    cat_df = df[df['category'] == category]
    
    for level in level_order:
        level_df = cat_df[cat_df['level'] == level]
        
        # Get control baseline
        control_scores = level_df[level_df['demographic_id'] == 'control']['score']
        if len(control_scores) == 0:
            continue
        control_mean = control_scores.mean()
        control_n = len(control_scores)
        
        # Store control row
        all_results.append({
            'category': category,
            'level': level,
            'level_label': level_labels.get(level, level),
            'demographic_id': 'control',
            'demographic_label': 'Control',
            'demographic_group': 'Control',
            'mean': control_mean,
            'delta': 0.0,
            'p_value': 1.0,
            'significance': '',
            'n': control_n
        })
        
        # Compute for each demographic
        for demo_id in demo_order:
            demo_scores = level_df[level_df['demographic_id'] == demo_id]['score']
            if len(demo_scores) == 0:
                continue
            
            demo_mean = demo_scores.mean()
            delta = demo_mean - control_mean
            
            # t-test
            if len(control_scores) > 1 and len(demo_scores) > 1:
                _, p_value = stats.ttest_ind(demo_scores, control_scores)
            else:
                p_value = 1.0
            
            sig = get_significance(p_value)
            
            all_results.append({
                'category': category,
                'level': level,
                'level_label': level_labels.get(level, level),
                'demographic_id': demo_id,
                'demographic_label': demo_labels.get(demo_id, demo_id),
                'demographic_group': demo_groups.get(demo_id, 'Other'),
                'mean': demo_mean,
                'delta': delta,
                'p_value': p_value,
                'significance': sig,
                'n': len(demo_scores)
            })

results_df = pd.DataFrame(all_results)

# === CREATE PIVOT TABLES FOR EACH CATEGORY ===
print("\nCreating formatted tables...")

from openpyxl import Workbook
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side
from openpyxl.utils import get_column_letter

wb = Workbook()

# Color definitions
header_fill = PatternFill(start_color="4472C4", end_color="4472C4", fill_type="solid")
header_font = Font(bold=True, color="FFFFFF", size=10)
control_fill = PatternFill(start_color="E2EFDA", end_color="E2EFDA", fill_type="solid")
group_header_fill = PatternFill(start_color="D9E2F3", end_color="D9E2F3", fill_type="solid")
sig_fill_3 = PatternFill(start_color="F4B183", end_color="F4B183", fill_type="solid")  # ***
sig_fill_2 = PatternFill(start_color="FFD966", end_color="FFD966", fill_type="solid")  # **
sig_fill_1 = PatternFill(start_color="FFF2CC", end_color="FFF2CC", fill_type="solid")  # *

thin_border = Border(
    left=Side(style='thin'),
    right=Side(style='thin'),
    top=Side(style='thin'),
    bottom=Side(style='thin')
)

def create_category_sheet(wb, category, results_df):
    """Create a detailed sheet for one question category"""
    ws = wb.create_sheet(title=f"{category.title()}")
    
    cat_data = results_df[results_df['category'] == category]
    
    # Title
    ws['A1'] = f"{category.upper()} QUESTIONS - Bias Analysis by Prompt Level"
    ws['A1'].font = Font(bold=True, size=14)
    ws.merge_cells('A1:K1')
    
    ws['A2'] = "Values shown: Mean score | Δ from control | Significance (* p<.05, ** p<.01, *** p<.001)"
    ws['A2'].font = Font(italic=True, size=9)
    ws.merge_cells('A2:K2')
    
    # Headers row
    start_row = 4
    ws.cell(row=start_row, column=1, value="Group").font = header_font
    ws.cell(row=start_row, column=1).fill = header_fill
    ws.cell(row=start_row, column=2, value="Demographic").font = header_font
    ws.cell(row=start_row, column=2).fill = header_fill
    
    for col_idx, level in enumerate(level_order, start=3):
        cell = ws.cell(row=start_row, column=col_idx, value=level_labels.get(level, level))
        cell.font = header_font
        cell.fill = header_fill
        cell.alignment = Alignment(horizontal='center', textRotation=45)
    
    # Control row
    row = start_row + 1
    ws.cell(row=row, column=1, value="Baseline").fill = control_fill
    ws.cell(row=row, column=2, value="Control").fill = control_fill
    
    for col_idx, level in enumerate(level_order, start=3):
        ctrl_data = cat_data[(cat_data['demographic_id'] == 'control') & (cat_data['level'] == level)]
        if len(ctrl_data) > 0:
            mean_val = ctrl_data['mean'].values[0]
            cell = ws.cell(row=row, column=col_idx, value=f"{mean_val:.2f}")
        else:
            cell = ws.cell(row=row, column=col_idx, value="--")
        cell.fill = control_fill
        cell.alignment = Alignment(horizontal='center')
        cell.border = thin_border
    
    # Demographics rows
    row += 1
    current_group = None
    
    for demo_id in demo_order:
        demo_group = demo_groups.get(demo_id, 'Other')
        demo_label = demo_labels.get(demo_id, demo_id)
        
        # Group separator
        if demo_group != current_group:
            if current_group is not None:
                row += 1  # Add spacing between groups
            current_group = demo_group
            # Group header row
            ws.cell(row=row, column=1, value=demo_group).font = Font(bold=True)
            ws.cell(row=row, column=1).fill = group_header_fill
            for col in range(2, len(level_order) + 3):
                ws.cell(row=row, column=col).fill = group_header_fill
            row += 1
        
        ws.cell(row=row, column=1, value="")
        ws.cell(row=row, column=2, value=demo_label)
        
        for col_idx, level in enumerate(level_order, start=3):
            demo_data = cat_data[(cat_data['demographic_id'] == demo_id) & (cat_data['level'] == level)]
            
            if len(demo_data) > 0:
                mean_val = demo_data['mean'].values[0]
                delta = demo_data['delta'].values[0]
                sig = demo_data['significance'].values[0]
                
                # Format: mean (delta) sig
                delta_str = f"+{delta:.2f}" if delta >= 0 else f"{delta:.2f}"
                cell_value = f"{mean_val:.2f}\n({delta_str}){sig}"
                
                cell = ws.cell(row=row, column=col_idx, value=cell_value)
                cell.alignment = Alignment(horizontal='center', vertical='center', wrap_text=True)
                cell.border = thin_border
                
                # Color based on significance
                if sig == '***':
                    cell.fill = sig_fill_3
                elif sig == '**':
                    cell.fill = sig_fill_2
                elif sig == '*':
                    cell.fill = sig_fill_1
            else:
                cell = ws.cell(row=row, column=col_idx, value="--")
                cell.alignment = Alignment(horizontal='center')
                cell.border = thin_border
        
        row += 1
    
    # Adjust column widths
    ws.column_dimensions['A'].width = 14
    ws.column_dimensions['B'].width = 22
    for col_idx in range(3, len(level_order) + 3):
        ws.column_dimensions[get_column_letter(col_idx)].width = 12
    
    # Set row heights for data rows
    for r in range(start_row + 2, row):
        ws.row_dimensions[r].height = 35
    
    return ws

# Create sheets for each category
for category in categories:
    create_category_sheet(wb, category, results_df)

# Remove default sheet
if 'Sheet' in wb.sheetnames:
    del wb['Sheet']

# === CREATE SUMMARY SHEET ===
ws_summary = wb.create_sheet(title="Summary", index=0)
ws_summary['A1'] = "BIAS ANALYSIS SUMMARY"
ws_summary['A1'].font = Font(bold=True, size=16)

ws_summary['A3'] = "This workbook contains detailed bias analysis across prompt complexity levels."
ws_summary['A4'] = "Each category sheet shows mean scores, deltas from control, and statistical significance."
ws_summary['A6'] = "Color Key:"
ws_summary['A7'] = "*** p < 0.001 (strong significance)"
ws_summary['A7'].fill = sig_fill_3
ws_summary['A8'] = "** p < 0.01"
ws_summary['A8'].fill = sig_fill_2
ws_summary['A9'] = "* p < 0.05"
ws_summary['A9'].fill = sig_fill_1

# Add overall statistics
ws_summary['A11'] = "Key Findings by Category:"
ws_summary['A11'].font = Font(bold=True, size=12)

row = 12
for category in categories:
    cat_data = results_df[(results_df['category'] == category) & (results_df['demographic_id'] != 'control')]
    sig_count = (cat_data['significance'] != '').sum()
    total = len(cat_data)
    avg_delta = cat_data['delta'].mean()
    max_bias = cat_data.loc[cat_data['delta'].abs().idxmax()]
    
    ws_summary[f'A{row}'] = f"{category.title()}:"
    ws_summary[f'A{row}'].font = Font(bold=True)
    row += 1
    ws_summary[f'A{row}'] = f"  - Significant effects: {sig_count}/{total} ({100*sig_count/total:.1f}%)"
    row += 1
    ws_summary[f'A{row}'] = f"  - Average delta: {avg_delta:+.3f}"
    row += 1
    ws_summary[f'A{row}'] = f"  - Largest bias: {max_bias['demographic_label']} at {max_bias['level_label']} (Δ={max_bias['delta']:+.2f})"
    row += 2

# === CREATE CROSS-CATEGORY COMPARISON ===
ws_cross = wb.create_sheet(title="Cross-Category")

ws_cross['A1'] = "CROSS-CATEGORY COMPARISON"
ws_cross['A1'].font = Font(bold=True, size=14)

# Create a table showing average bias by level across all categories
ws_cross['A3'] = "Average Δ from Control by Prompt Level and Category"
ws_cross['A3'].font = Font(bold=True)

start_row = 5
ws_cross.cell(row=start_row, column=1, value="Prompt Level").font = header_font
ws_cross.cell(row=start_row, column=1).fill = header_fill

for col_idx, cat in enumerate(categories, start=2):
    ws_cross.cell(row=start_row, column=col_idx, value=cat.title()).font = header_font
    ws_cross.cell(row=start_row, column=col_idx).fill = header_fill

row = start_row + 1
for level in level_order:
    ws_cross.cell(row=row, column=1, value=level_labels.get(level, level))
    for col_idx, cat in enumerate(categories, start=2):
        level_cat_data = results_df[
            (results_df['level'] == level) & 
            (results_df['category'] == cat) & 
            (results_df['demographic_id'] != 'control')
        ]
        if len(level_cat_data) > 0:
            avg_delta = level_cat_data['delta'].mean()
            sig_pct = (level_cat_data['significance'] != '').mean() * 100
            cell = ws_cross.cell(row=row, column=col_idx, value=f"{avg_delta:+.3f}\n({sig_pct:.0f}% sig)")
            cell.alignment = Alignment(horizontal='center', wrap_text=True)
            if abs(avg_delta) > 0.15:
                cell.fill = sig_fill_2
            elif abs(avg_delta) > 0.1:
                cell.fill = sig_fill_1
    row += 1

ws_cross.column_dimensions['A'].width = 16
for col_idx in range(2, len(categories) + 2):
    ws_cross.column_dimensions[get_column_letter(col_idx)].width = 14

# Add demographic comparison across categories
ws_cross.cell(row=row + 2, column=1, value="Average Δ by Demographic Group Across Categories").font = Font(bold=True)

start_row2 = row + 4
ws_cross.cell(row=start_row2, column=1, value="Demographic").font = header_font
ws_cross.cell(row=start_row2, column=1).fill = header_fill

for col_idx, cat in enumerate(categories, start=2):
    ws_cross.cell(row=start_row2, column=col_idx, value=cat.title()).font = header_font
    ws_cross.cell(row=start_row2, column=col_idx).fill = header_fill

row = start_row2 + 1
for demo_id in demo_order:
    ws_cross.cell(row=row, column=1, value=demo_labels.get(demo_id, demo_id))
    for col_idx, cat in enumerate(categories, start=2):
        demo_cat_data = results_df[
            (results_df['demographic_id'] == demo_id) & 
            (results_df['category'] == cat)
        ]
        if len(demo_cat_data) > 0:
            avg_delta = demo_cat_data['delta'].mean()
            sig_count = (demo_cat_data['significance'] != '').sum()
            cell = ws_cross.cell(row=row, column=col_idx, value=f"{avg_delta:+.3f} ({sig_count})")
            cell.alignment = Alignment(horizontal='center')
            if avg_delta < -0.1:
                cell.font = Font(color="C00000")  # Red for negative
            elif avg_delta > 0.1:
                cell.font = Font(color="006400")  # Green for positive
    row += 1


for category in categories:
    print(f"\n{'='*80}")
    print(f"CATEGORY: {category.upper()}")
    print(f"{'='*80}")
    
    # Create pivot for this category
    cat_data = results_df[results_df['category'] == category].copy()
    
    # Control row
    control_pivot = cat_data[cat_data['demographic_id'] == 'control'].pivot(
        index='demographic_id', columns='level', values='mean'
    )
    control_pivot = control_pivot.reindex(columns=level_order)
    print("\nControl baseline:")
    print(control_pivot.round(2).to_string())
    
    # Delta pivot
    demo_data = cat_data[cat_data['demographic_id'] != 'control'].copy()
    
    # Create combined mean/delta/sig string
    def format_cell(row):
        sig = row['significance']
        return f"{row['mean']:.2f} ({row['delta']:+.2f}){sig}"
    
    demo_data['formatted'] = demo_data.apply(format_cell, axis=1)
    
    formatted_pivot = demo_data.pivot(
        index='demographic_id', columns='level', values='formatted'
    )
    formatted_pivot = formatted_pivot.reindex(index=demo_order, columns=level_order)
    
    print("\nDemographic results (mean (Δ)sig):")
    print(formatted_pivot.to_string())

# === CREATE VISUALIZATION ===
print("\n\nCreating visualizations...")

import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

fig, axes = plt.subplots(3, 1, figsize=(14, 16))

for idx, category in enumerate(categories):
    ax = axes[idx]
    
    cat_data = results_df[(results_df['category'] == category) & (results_df['demographic_id'] != 'control')]
    
    # Pivot for heatmap
    pivot = cat_data.pivot_table(index='demographic_label', columns='level', values='delta', aggfunc='mean')
    pivot = pivot.reindex(columns=level_order)
    pivot = pivot.reindex([demo_labels[d] for d in demo_order if demo_labels[d] in pivot.index])
    
    # Heatmap
    im = ax.imshow(pivot.values, cmap='RdBu_r', aspect='auto', vmin=-0.4, vmax=0.4)
    
    ax.set_xticks(range(len(level_order)))
    ax.set_xticklabels([level_labels[l] for l in level_order], rotation=45, ha='right', fontsize=9)
    ax.set_yticks(range(len(pivot.index)))
    ax.set_yticklabels(pivot.index, fontsize=9)
    
    ax.set_title(f"{category.upper()} Questions: Δ from Control by Demographic × Prompt Level", 
                 fontsize=12, fontweight='bold', pad=10)
    
    # Add values
    for i in range(len(pivot.index)):
        for j in range(len(pivot.columns)):
            val = pivot.iloc[i, j]
            if pd.notna(val):
                color = 'white' if abs(val) > 0.2 else 'black'
                ax.text(j, i, f'{val:+.2f}', ha='center', va='center', color=color, fontsize=7)
    
    plt.colorbar(im, ax=ax, label="Δ from Control", shrink=0.8)

plt.tight_layout()
plt.show()

# === ADDITIONAL: BAR CHART COMPARISON ===
fig2, axes2 = plt.subplots(1, 3, figsize=(18, 8))

for idx, category in enumerate(categories):
    ax = axes2[idx]
    
    cat_data = results_df[(results_df['category'] == category) & (results_df['demographic_id'] != 'control')]
    
    # Average delta by level
    level_avg = cat_data.groupby('level')['delta'].mean().reindex(level_order)
    level_sig = cat_data.groupby('level').apply(lambda x: (x['significance'] != '').mean() * 100).reindex(level_order)
    
    colors = ['#FF6B6B' if d < -0.05 else '#4ECDC4' if d > 0.05 else '#95A5A6' for d in level_avg]
    
    bars = ax.bar(range(len(level_order)), level_avg, color=colors, edgecolor='black', linewidth=0.5)
    
    # Add significance % labels
    for i, (bar, sig) in enumerate(zip(bars, level_sig)):
        height = bar.get_height()
        ax.annotate(f'{sig:.0f}%', xy=(bar.get_x() + bar.get_width()/2, height),
                   xytext=(0, 3), textcoords="offset points",
                   ha='center', va='bottom', fontsize=8)
    
    ax.set_xticks(range(len(level_order)))
    ax.set_xticklabels([level_labels[l] for l in level_order], rotation=45, ha='right', fontsize=9)
    ax.set_ylabel("Average Δ from Control")
    ax.set_title(f"{category.upper()}", fontsize=12, fontweight='bold')
    ax.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
    ax.axhline(y=0.1, color='green', linestyle='--', alpha=0.5, linewidth=0.5)
    ax.axhline(y=-0.1, color='red', linestyle='--', alpha=0.5, linewidth=0.5)

fig2.suptitle("Average Bias by Prompt Level (% = proportion significant)", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
"""
Bias Analysis Visualizations - Notebook Version
Run cells individually to display each visualization inline
"""

import pandas as pd
import numpy as np
from scipy import stats
import json
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.lines import Line2D
from math import pi

# === LOAD DATA ===
df = pd.read_csv('../results_raw.csv')
with open('../prompts/demographics.json', 'r') as f:
    demographics = json.load(f)

# === CONFIG ===
level_order = ['minimal', 'context_only', 'vague_positive', 'vague_negative',
               'neutral_metrics', 'positive_metrics', 'negative_metrics',
               'contradict_pos_metrics', 'contradict_neg_metrics']

level_labels = {'minimal': 'Minimal', 'context_only': 'Context', 'vague_positive': 'Vague+',
                'vague_negative': 'Vague-', 'neutral_metrics': 'Neutral', 'positive_metrics': 'Positive',
                'negative_metrics': 'Negative', 'contradict_pos_metrics': 'Contra+', 'contradict_neg_metrics': 'Contra-'}

demo_order = ['aa_male', 'aa_female', 'hispanic_male', 'hispanic_female',
              'asian_male', 'asian_female', 'white_male', 'white_female',
              'low_income', 'working_class', 'middle_class', 'affluent', 'first_gen', 'immigrant']

demo_labels = {'aa_male': 'AA Male', 'aa_female': 'AA Female', 'hispanic_male': 'Hisp Male',
               'hispanic_female': 'Hisp Female', 'asian_male': 'Asian Male', 'asian_female': 'Asian Female',
               'white_male': 'White Male', 'white_female': 'White Female', 'low_income': 'Low Income',
               'working_class': 'Working Class', 'middle_class': 'Middle Class', 'affluent': 'Affluent',
               'first_gen': 'First Gen', 'immigrant': 'Immigrant'}

demo_groups = {'aa_male': 'African American', 'aa_female': 'African American',
               'hispanic_male': 'Hispanic', 'hispanic_female': 'Hispanic',
               'asian_male': 'Asian', 'asian_female': 'Asian',
               'white_male': 'White', 'white_female': 'White',
               'low_income': 'Socioeconomic', 'working_class': 'Socioeconomic',
               'middle_class': 'Socioeconomic', 'affluent': 'Socioeconomic',
               'first_gen': 'Background', 'immigrant': 'Background'}

group_colors = {'African American': '#E64B35', 'Hispanic': '#4DBBD5', 'Asian': '#00A087',
                'White': '#3C5488', 'Socioeconomic': '#F39B7F', 'Background': '#8491B4'}

categories = ['opportunity', 'potential', 'discipline']

# === COMPUTE RESULTS ===
def compute_results():
    results = []
    for category in categories:
        cat_df = df[df['category'] == category]
        for level in level_order:
            level_df = cat_df[cat_df['level'] == level]
            control = level_df[level_df['demographic_id'] == 'control']['score']
            if len(control) == 0: continue
            control_mean, control_std, control_n = control.mean(), control.std(), len(control)
            
            for demo_id in demo_order:
                demo = level_df[level_df['demographic_id'] == demo_id]['score']
                if len(demo) == 0: continue
                delta = demo.mean() - control_mean
                se = np.sqrt(demo.std()**2/len(demo) + control_std**2/control_n)
                _, p = stats.ttest_ind(demo, control)
                results.append({
                    'category': category, 'level': level, 'level_idx': level_order.index(level),
                    'demographic_id': demo_id, 'demographic_label': demo_labels[demo_id],
                    'demographic_group': demo_groups[demo_id], 'mean': demo.mean(), 'delta': delta,
                    'se': se, 'ci_lower': delta - 1.96*se, 'ci_upper': delta + 1.96*se,
                    'p_value': p, 'significant': p < 0.05
                })
    return pd.DataFrame(results)

results_df = compute_results()

#%% ============================================================================
# VIZ 1: HEATMAP WITH SIGNIFICANCE MARKERS
# ==============================================================================
fig, axes = plt.subplots(1, 3, figsize=(18, 10))
colors_neg = plt.cm.Blues_r(np.linspace(0.2, 0.8, 128))
colors_pos = plt.cm.Reds(np.linspace(0.2, 0.8, 128))
cmap = LinearSegmentedColormap.from_list('diverging', np.vstack([colors_neg, colors_pos]))

for idx, category in enumerate(categories):
    ax = axes[idx]
    cat_data = results_df[results_df['category'] == category]
    pivot_delta = cat_data.pivot_table(index='demographic_id', columns='level', values='delta').reindex(index=demo_order, columns=level_order)
    pivot_sig = cat_data.pivot_table(index='demographic_id', columns='level', values='significant').reindex(index=demo_order, columns=level_order)
    
    im = ax.imshow(pivot_delta.values, cmap=cmap, aspect='auto', vmin=-0.5, vmax=0.5)
    for i in range(len(demo_order)):
        for j in range(len(level_order)):
            val, sig = pivot_delta.iloc[i, j], pivot_sig.iloc[i, j]
            if pd.notna(val):
                color = 'white' if abs(val) > 0.25 else 'black'
                ax.text(j, i, f'{val:+.2f}{"•" if sig else ""}', ha='center', va='center', 
                       color=color, fontsize=7, fontweight='bold' if sig else 'normal')
    
    for y in [1.5, 3.5, 5.5, 7.5, 11.5]: ax.axhline(y=y, color='white', linewidth=2)
    ax.set_xticks(range(len(level_order))); ax.set_xticklabels([level_labels[l] for l in level_order], rotation=45, ha='right', fontsize=9)
    ax.set_yticks(range(len(demo_order))); ax.set_yticklabels([demo_labels[d] for d in demo_order], fontsize=9)
    ax.set_title(f'{category.upper()}', fontsize=14, fontweight='bold')

fig.colorbar(im, ax=axes, orientation='vertical', fraction=0.02, pad=0.02, label='Δ from Control')
fig.suptitle('Bias by Demographic and Prompt Level (• = p < 0.05)', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

#%% ============================================================================
# VIZ 2: FOREST PLOT WITH CONFIDENCE INTERVALS
# ==============================================================================
fig, axes = plt.subplots(1, 3, figsize=(16, 12))
for idx, category in enumerate(categories):
    ax = axes[idx]
    demo_avg = results_df[results_df['category'] == category].groupby(['demographic_id', 'demographic_group']).agg(
        {'delta': 'mean', 'ci_lower': 'mean', 'ci_upper': 'mean', 'significant': 'sum'}).reset_index()
    demo_avg = demo_avg.set_index('demographic_id').reindex(demo_order).reset_index()
    
    for i, row in demo_avg.iterrows():
        color = group_colors[row['demographic_group']]
        ax.plot([row['ci_lower'], row['ci_upper']], [i, i], color=color, linewidth=2, alpha=0.6)
        ax.scatter(row['delta'], i, color=color, s=80, marker='o' if row['significant'] >= 3 else 's', 
                  zorder=5, edgecolor='black', linewidth=0.5)
    
    ax.axvline(x=0, color='black', linestyle='-', linewidth=1, alpha=0.5)
    ax.set_yticks(range(len(demo_order))); ax.set_yticklabels([demo_labels[d] for d in demo_order], fontsize=10)
    ax.set_xlabel('Average Δ from Control'); ax.set_title(f'{category.upper()}', fontsize=14, fontweight='bold')
    ax.set_xlim(-0.6, 0.6); ax.invert_yaxis(); ax.grid(axis='x', alpha=0.3)

fig.legend(handles=[mpatches.Patch(facecolor=c, label=g) for g, c in group_colors.items()], 
           loc='lower center', ncol=6, fontsize=10, bbox_to_anchor=(0.5, -0.02))
fig.suptitle('Average Bias with 95% CI (○ = 3+ significant levels)', fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()

#%% ============================================================================
# VIZ 3: TRAJECTORY LINES ACROSS PROMPT LEVELS
# ==============================================================================
fig, axes = plt.subplots(3, 2, figsize=(16, 14))
race_demos = ['aa_male', 'aa_female', 'hispanic_male', 'hispanic_female', 'asian_male', 'asian_female', 'white_male', 'white_female']
ses_demos = ['low_income', 'working_class', 'middle_class', 'affluent', 'first_gen', 'immigrant']
ses_colors = {'low_income': '#d62728', 'working_class': '#ff7f0e', 'middle_class': '#2ca02c', 
              'affluent': '#1f77b4', 'first_gen': '#9467bd', 'immigrant': '#8c564b'}

for row_idx, category in enumerate(categories):
    cat_data = results_df[results_df['category'] == category]
    
    # Race/Gender
    ax = axes[row_idx, 0]
    for demo_id in race_demos:
        d = cat_data[cat_data['demographic_id'] == demo_id].sort_values('level_idx')
        ax.plot(d['level_idx'], d['delta'], color=group_colors[demo_groups[demo_id]], 
               linestyle='-' if 'male' in demo_id else '--', linewidth=2, marker='o', markersize=5, 
               label=demo_labels[demo_id], alpha=0.8)
    ax.axhline(y=0, color='black', linewidth=1); ax.set_ylim(-0.5, 0.7)
    ax.set_xticks(range(len(level_order))); ax.set_xticklabels([level_labels[l] for l in level_order], rotation=45, ha='right', fontsize=8)
    ax.set_ylabel('Δ from Control'); ax.set_title(f'{category.upper()} - Race/Gender', fontsize=12, fontweight='bold'); ax.grid(alpha=0.3)
    if row_idx == 0: ax.legend(loc='upper right', fontsize=7, ncol=2)
    
    # SES
    ax = axes[row_idx, 1]
    for demo_id in ses_demos:
        d = cat_data[cat_data['demographic_id'] == demo_id].sort_values('level_idx')
        ax.plot(d['level_idx'], d['delta'], color=ses_colors[demo_id], linewidth=2, marker='s', markersize=5, 
               label=demo_labels[demo_id], alpha=0.8)
    ax.axhline(y=0, color='black', linewidth=1); ax.set_ylim(-1.0, 1.2)
    ax.set_xticks(range(len(level_order))); ax.set_xticklabels([level_labels[l] for l in level_order], rotation=45, ha='right', fontsize=8)
    ax.set_ylabel('Δ from Control'); ax.set_title(f'{category.upper()} - Socioeconomic', fontsize=12, fontweight='bold'); ax.grid(alpha=0.3)
    if row_idx == 0: ax.legend(loc='upper right', fontsize=8)

fig.suptitle('Bias Trajectory Across Prompt Complexity Levels', fontsize=16, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()

#%% ============================================================================
# VIZ 4: DIVERGING BAR CHART
# ==============================================================================
fig, axes = plt.subplots(3, 1, figsize=(14, 16))
for idx, category in enumerate(categories):
    ax = axes[idx]
    demo_avg = results_df[results_df['category'] == category].groupby('demographic_id')['delta'].mean().reindex(demo_order)
    colors = [group_colors[demo_groups[d]] for d in demo_order]
    bars = ax.barh(range(len(demo_order)), demo_avg.values, color=colors, edgecolor='black', linewidth=0.5, height=0.7)
    for i, (bar, val) in enumerate(zip(bars, demo_avg.values)):
        ax.text(val + 0.02 if val >= 0 else val - 0.02, i, f'{val:+.2f}', va='center', 
               ha='left' if val >= 0 else 'right', fontsize=9, fontweight='bold')
    ax.axvline(x=0, color='black', linewidth=1)
    ax.set_yticks(range(len(demo_order))); ax.set_yticklabels([demo_labels[d] for d in demo_order], fontsize=10)
    ax.set_xlabel('Average Δ from Control'); ax.set_title(f'{category.upper()}', fontsize=14, fontweight='bold')
    ax.set_xlim(-0.5, 0.5); ax.grid(axis='x', alpha=0.3); ax.invert_yaxis()

fig.legend(handles=[mpatches.Patch(facecolor=c, label=g, edgecolor='black') for g, c in group_colors.items()],
           loc='lower center', ncol=6, fontsize=10, bbox_to_anchor=(0.5, -0.01))
fig.suptitle('Average Bias by Demographic', fontsize=16, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()

#%% ============================================================================
# VIZ 5: DOT MATRIX (Size = magnitude, Color = direction, Border = sig)
# ==============================================================================
fig, axes = plt.subplots(1, 3, figsize=(18, 10))
for idx, category in enumerate(categories):
    ax = axes[idx]
    cat_data = results_df[results_df['category'] == category]
    pivot = cat_data.pivot_table(index='demographic_id', columns='level', values='delta').reindex(index=demo_order, columns=level_order)
    pivot_sig = cat_data.pivot_table(index='demographic_id', columns='level', values='significant').reindex(index=demo_order, columns=level_order)
    
    for i, demo_id in enumerate(demo_order):
        for j, level in enumerate(level_order):
            val = pivot.loc[demo_id, level] if demo_id in pivot.index else np.nan
            sig = pivot_sig.loc[demo_id, level] if demo_id in pivot_sig.index else False
            if pd.notna(val):
                ax.scatter(j, i, s=abs(val)*400+20, c='#d62728' if val < 0 else '#2ca02c',
                          alpha=0.9 if sig else 0.4, edgecolors='black' if sig else 'none', linewidths=1.5)
    
    ax.set_xticks(range(len(level_order))); ax.set_xticklabels([level_labels[l] for l in level_order], rotation=45, ha='right', fontsize=9)
    ax.set_yticks(range(len(demo_order))); ax.set_yticklabels([demo_labels[d] for d in demo_order], fontsize=9)
    ax.set_title(f'{category.upper()}', fontsize=14, fontweight='bold'); ax.invert_yaxis(); ax.grid(alpha=0.2)

fig.legend(handles=[Line2D([0], [0], marker='o', color='w', markerfacecolor='#2ca02c', markersize=10, label='Positive Δ'),
                    Line2D([0], [0], marker='o', color='w', markerfacecolor='#d62728', markersize=10, label='Negative Δ'),
                    Line2D([0], [0], marker='o', color='w', markerfacecolor='gray', markeredgecolor='black', markeredgewidth=2, markersize=10, label='Significant')],
           loc='lower center', ncol=5, fontsize=10, bbox_to_anchor=(0.5, -0.02))
fig.suptitle('Bias: Size = |Δ|, Color = Direction, Border = Significant', fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()

#%% ============================================================================
# VIZ 6: SMALL MULTIPLES (One panel per demographic)
# ==============================================================================
fig, axes = plt.subplots(4, 4, figsize=(18, 16))
axes = axes.flatten()
cat_colors = {'opportunity': '#1f77b4', 'potential': '#2ca02c', 'discipline': '#d62728'}

for i, demo_id in enumerate(demo_order):
    ax = axes[i]
    for category in categories:
        d = results_df[(results_df['category'] == category) & (results_df['demographic_id'] == demo_id)].sort_values('level_idx')
        ax.plot(d['level_idx'], d['delta'], color=cat_colors[category], linewidth=2, marker='o', markersize=4, label=category.title())
        sig_d = d[d['significant']]
        ax.scatter(sig_d['level_idx'], sig_d['delta'], color=cat_colors[category], s=60, zorder=5, edgecolor='black', linewidth=1)
    ax.axhline(y=0, color='black', linewidth=0.5); ax.set_xlim(-0.5, 8.5); ax.set_ylim(-1.0, 1.2)
    ax.set_title(demo_labels[demo_id], fontsize=11, fontweight='bold'); ax.set_xticks(range(len(level_order)))
    ax.set_xticklabels([] if i < 12 else [level_labels[l][:3] for l in level_order], rotation=45, ha='right', fontsize=7)
    if i % 4 == 0: ax.set_ylabel('Δ', fontsize=10)
    ax.grid(alpha=0.2)

for ax in axes[14:]: ax.axis('off')
axes[14].legend(handles=[Line2D([0], [0], color=c, linewidth=2, marker='o', label=cat.title()) for cat, c in cat_colors.items()], loc='center', fontsize=12)
fig.suptitle('Bias Trajectories by Demographic', fontsize=16, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()

#%% ============================================================================
# VIZ 7: SLOPE CHART (Minimal → Positive Metrics)
# ==============================================================================
fig, axes = plt.subplots(1, 3, figsize=(15, 10))
for idx, category in enumerate(categories):
    ax = axes[idx]
    cat_data = results_df[results_df['category'] == category]
    for demo_id in demo_order:
        d = cat_data[cat_data['demographic_id'] == demo_id]
        v1 = d[d['level'] == 'minimal']['delta'].values
        v2 = d[d['level'] == 'positive_metrics']['delta'].values
        if len(v1) and len(v2):
            color = group_colors[demo_groups[demo_id]]
            ax.plot([0, 1], [v1[0], v2[0]], color=color, linewidth=2, alpha=0.7)
            ax.scatter([0, 1], [v1[0], v2[0]], color=color, s=60, zorder=5, edgecolor='black', linewidth=0.5)
            ax.text(1.05, v2[0], demo_labels[demo_id], va='center', fontsize=8, color=color)
    ax.set_xlim(-0.2, 1.5); ax.set_xticks([0, 1]); ax.set_xticklabels(['Minimal\n(No metrics)', 'Positive Metrics\n(Explicit data)'], fontsize=11)
    ax.set_ylabel('Δ from Control'); ax.set_title(f'{category.upper()}', fontsize=14, fontweight='bold')
    ax.axhline(y=0, color='black', linewidth=0.5); ax.grid(axis='y', alpha=0.3)

fig.suptitle('How Explicit Performance Metrics Change Bias', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

#%% ============================================================================
# VIZ 8: RADAR CHART BY DEMOGRAPHIC GROUP
# ==============================================================================
fig, axes = plt.subplots(2, 3, figsize=(15, 12), subplot_kw=dict(polar=True))
axes = axes.flatten()
grouped_demos = {'African American': ['aa_male', 'aa_female'], 'Hispanic': ['hispanic_male', 'hispanic_female'],
                 'Asian': ['asian_male', 'asian_female'], 'White': ['white_male', 'white_female'],
                 'Lower SES': ['low_income', 'working_class'], 'Higher SES': ['middle_class', 'affluent']}
radar_levels = ['minimal', 'context_only', 'vague_negative', 'neutral_metrics', 'positive_metrics', 'negative_metrics']
angles = [n / float(len(radar_levels)) * 2 * pi for n in range(len(radar_levels))] + [0]
cat_colors = {'opportunity': '#1f77b4', 'potential': '#2ca02c', 'discipline': '#d62728'}

for ax_idx, (group_name, demo_ids) in enumerate(grouped_demos.items()):
    ax = axes[ax_idx]
    for category in categories:
        level_avg = results_df[(results_df['category'] == category) & (results_df['demographic_id'].isin(demo_ids))].groupby('level')['delta'].mean()
        values = [level_avg.get(l, 0) for l in radar_levels] + [level_avg.get(radar_levels[0], 0)]
        ax.plot(angles, values, color=cat_colors[category], linewidth=2, label=category.title())
        ax.fill(angles, values, color=cat_colors[category], alpha=0.1)
    ax.set_xticks(angles[:-1]); ax.set_xticklabels([level_labels[l] for l in radar_levels], fontsize=8)
    ax.set_title(group_name, fontsize=12, fontweight='bold', pad=20); ax.set_ylim(-0.6, 0.6)
    if ax_idx == 0: ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0), fontsize=9)

fig.suptitle('Bias Profile by Demographic Group', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```
